```python
# import Pandas and read school csv
import pandas as pd
school_file = "Pandas_Data/schools_complete.csv"
school_file_df = pd.read_csv(school_file)
# school_file_df.head()
```

```python
# Total number of Schools, Students, and budget 
total_num_schools = len("school_name")
total_students = school_file_df["size"].sum()
total_budget = school_file_df["budget"].sum()
student_count = school_file_df.groupby(["school_name"])["size"].max()
```

```python
# Budget per Student 
school_file_df["budget_student"] = school_file_df["budget"] / school_file_df["size"]
```

```python
# Read Students CSV
student_file = "Pandas_Data/students_complete.csv"
student_file_df = pd.read_csv(student_file)

# student_file_df.head()
```

```python
# Average District score for math & reading
average_math = student_file_df["math_score"].mean()
average_read = student_file_df["reading_score"].mean()
```

```python
# District percentage passing reading
passing_read_df = student_file_df.loc[student_file_df["reading_score"] >= 70]
percent_passing_read_count = passing_read_df["reading_score"].count()
percent_passing_read = percent_passing_read_count/total_students
# percent_passing_read
```

```python
# District percentage passing math
passing_math_df = student_file_df.loc[student_file_df["math_score"] >= 70]
percent_passing_math_count = passing_math_df["math_score"].count()
percent_passing_math = percent_passing_math_count/total_students 
# percent_passing_math
```

```python
# District percentage passing reading
overall_passing = (percent_passing_math + percent_passing_read)/2
# overall_passing
```

```python
# Disctrict Summary Print-Out
District_Summary = pd.DataFrame({"Total Number of Schools":[total_num_schools],
                                "Total Number of Students":[total_students],
                                "Total Budget":[total_budget],
                                "Average Math Score":[average_math],
                                "Average Reading Score":[average_read],
                                "Passing Math":[percent_passing_math],
                                "Passing Reading":[percent_passing_read],
                                "Overall Passing":[overall_passing]})
District_Summary
```

```python
# Merged DataFrames
school_student_df = pd.merge(student_file_df,school_file_df, how = "left", on = ["school_name","school_name"])
# school_student_df.head()
```

```python
# Per School Calculations/pre-printouts
student_count_hs = school_student_df["school_name"].value_counts()
type_hs = school_student_df.groupby(["school_name"])["type"].max()
school_budget = school_student_df.groupby(["school_name"]).mean()["budget"]
perschool_spending = school_budget/student_count
mean_math_score = school_student_df.groupby(["school_name"]).mean()["math_score"]
mean_read_score = school_student_df.groupby(["school_name"]).mean()["reading_score"]
```

```python
# % passing reading by School
passing_read_df_hs = school_student_df.loc[student_file_df["reading_score"] >= 70]
percent_passing_read_hs_count = school_student_df.groupby(["school_name"]).mean()["reading_score"]
# percent_passing_read_hs_count = passing_read_df_hs["reading_score"].count()
# percent_passing_read_hs = school_student_df.groupby(["school_name"])["size"].max()
```

```python
# # % passing math by School
passing_math_df_hs = school_student_df.loc[school_student_df["math_score"] >= 70]
percent_passing_math_hs_count = school_student_df.groupby(["school_name"]).mean()["math_score"]
# # percent_passing_math_hs_count = passing_math_df_hs["math_score"].count()
# # percent_passing_math_hs = school_student_df.groupby(["school_name"])["size"].max()
# percent_passing_math_hs_count
```

```python
# Both passing percentage
both_passing_percentage = (percent_passing_math_hs_count + percent_passing_read_hs_count)/2
# both_passing_percentage
```

```python
# General print out for by school stats
Schools_Summary = pd.DataFrame({"School Type":type_hs,
                                "Student Count":student_count_hs,
                                "School Budget":school_budget,
                                "Per Student Budget":perschool_spending,
                                "Average Math Score":mean_math_score,
                                "Average Reading Score":mean_read_score,
                                "School Passing Reading":percent_passing_read_hs_count,
                                "School Passing Math":percent_passing_math_hs_count,
                                "School Overall Passing":both_passing_percentage})
Schools_Summary
```
